```python
# Handling missing data
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', None],
    'Age': [24, np.nan, 22, 32, np.nan, 29],
    'City': ['Delhi', 'Mumbai', None, 'Kolkata', 'Hyderabad', 'Chennai'],
    'Salary': [50000, 60000, 55000, None, 52000, 58000]
}
```

```python
data_frame = pd.DataFrame(data)
```

```python
data_frame
```

```python
# jaha par null value hai vaha True hai ar jaha par null value nai hai vaha false hai
data_frame.isnull()
```

```python
# jaha par null values hai vaha false aur jaha pr values hai vaha true hai
data_frame.notnull()
```

```python
data_frame.fillna('Missing values')
```

```python
data_frame.dropna()
```
