```python
#Create a DataFrame with sales data: Product, Price, Quantity Sold, Total Sales (Calculate as Price *
#Quantity Sold)
import pandas as pd
```

```python
data = {
    'Products':["Laptop", "Mouse", "Keyboard", "Processor", "Monitor", "Web cam"],
    'Price' : [60000, 1200, 1300, 2000, 5000, 500],
    'Quantity_sold' : [10, 15, 20, 25, 40, 60],
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Total_Sales'] = df['Price'] * df['Quantity_sold']
df
```

```python
# Add a new column Discounted Price where discount = 10% on price
df['Discounted_Price'] = df['Price'] * 0.90
df
```

```python
# Filter the products where Total Sales > 500.
filtered_products = df[df['Total_Sales'] > 100000]
print(filtered_products)
```
